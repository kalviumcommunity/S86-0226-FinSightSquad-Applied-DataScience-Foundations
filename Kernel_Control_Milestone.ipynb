# Jupyter Kernel Control & Debugging Milestone
This notebook demonstrates how to control the Jupyter kernel, run cells in sequence, restart the kernel, and interrupt execution.

## 1. Running Cells and Understanding Execution Order
The kernel remembers variables and their values until restarted.

```python
a = 10
print('a =', a)
```

```python
b = a + 5
print('b =', b)
```

## 2. Restarting the Kernel
Restart the kernel using the menu. After restart, variables are cleared. Try running the next cell before rerunning previous cells.

```python
print(a)  # This will cause an error after restart until you rerun the cell that defines 'a'.
```

## 3. Interrupting Execution
Start a long-running operation and interrupt it using the interrupt option.

```python
import time
for i in range(100):
    print(i)
    time.sleep(0.5)  # Interrupt this cell while it is running.
```

## 4. Restart vs Interrupt
- Use interrupt for stopping a single stuck cell.
- Use restart to clear all variables and reset the notebook state.
- Restart is safer when the kernel is unresponsive or you want a clean slate.

## 5. Summary
- Run cells in order for predictable results.
- Restart kernel to clear memory.
- Interrupt to stop long-running code.
- Always restart and rerun before final submission.
