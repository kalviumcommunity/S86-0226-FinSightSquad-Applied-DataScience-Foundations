# Understanding Notebook Cells: Code vs Markdown

This notebook demonstrates how to separate **execution** (Code cells) from **explanation** (Markdown cells).

## Notebook Purpose

By the end of this notebook, you should be able to:
- identify when to use Code cells
- identify when to use Markdown cells
- build a clear notebook structure that others can read

## Example 1: Code Cell (Execution)

The next cell runs Python code and returns output.

```python
message = "This is a Code cell. It executes Python."
print(message)
```

### What happened?

The previous Code cell executed Python and displayed text output.
This is what Code cells are for: calculations, logic, and execution.

## Example 2: Another Code Cell

The next cell performs a simple computation.

```python
a = 8
b = 5
total = a + b
print("Sum:", total)
```

### Why this is structured well

- Markdown explains **what** will happen and **why** it matters.
- Code performs the actual computation.
- This separation makes notebooks easier to review and debug.
